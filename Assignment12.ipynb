```python
class Money:
    def __init__(self, value):
        self.value = value

    def __str__(self):
        return f"${self.value:.2f}"

class Bills(Money):
    pass

class Coins(Money):
    pass

class One(Bills):
    def __init__(self):
        super().__init__(1.0)

class Five(Bills):
    def __init__(self):
        super().__init__(5.0)

class Ten(Bills):
    def __init__(self):
        super().__init__(10.0)

class Twenty(Bills):
    def __init__(self):
        super().__init__(20.0)

class Penny(Coins):
    def __init__(self):
        super().__init__(0.01)

class Nickel(Coins):
    def __init__(self):
        super().__init__(0.05)

class Dime(Coins):
    def __init__(self):
        super().__init__(0.10)

class Quarter(Coins):
    def __init__(self):
        super().__init__(0.25)

class Wallet:
    def __init__(self, money=[]):
        self.money = money

    def add_money(self, money):
        self.money.extend(money)

    def total_value(self):
        return sum([item.value for item in self.money])

    def __str__(self):
        money_str = ", ".join([str(item) for item in self.money])
        return f"Wallet: {money_str}"

# Test cases
if __name__ == "__main__":

    one_dollar = One()
    five_dollar = Five()
    ten_dollar = Ten()
    twenty_dollar = Twenty()

    penny = Penny()
    nickel = Nickel()
    dime = Dime()
    quarter = Quarter()

    wallet1 = Wallet([one_dollar, five_dollar, penny, dime])
    wallet2 = Wallet([twenty_dollar, nickel, quarter])
    wallet3 = Wallet([ten_dollar, dime, dime, quarter])

    wallet1.add_money([ten_dollar, quarter])
    wallet2.add_money([one_dollar, five_dollar])
    wallet3.add_money([one_dollar, one_dollar, quarter])

    print(wallet1)
    print("Total value:", wallet1.total_value())

    print(wallet2)
    print("Total value:", wallet2.total_value())

    print(wallet3)
    print("Total value:", wallet3.total_value())
```

```python
# design to represent different types of vehicles and garage that holds them

class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def __str__(self):
        return f"{self.brand} {self.model}"

class Car(Vehicle):
    def __init__(self, brand, model, num_doors):
        super().__init__(brand, model)
        self.num_doors = num_doors

class Truck(Vehicle):
    def __init__(self, brand, model, cargo_capacity):
        super().__init__(brand, model)
        self.cargo_capacity = cargo_capacity

class Motorcycle(Vehicle):
    def __init__(self, brand, model, has_sidecar=False):
        super().__init__(brand, model)
        self.has_sidecar = has_sidecar

class Garage:
    def __init__(self, vehicles=[]):
        self.vehicles = vehicles

    def add_vehicle(self, vehicle):
        self.vehicles.append(vehicle)

    def __str__(self):
        vehicle_str = "\n".join([str(vehicle) for vehicle in self.vehicles])
        return f"Garage:\n{vehicle_str}"

# Test cases
if __name__ == "__main__":

    car1 = Car("Toyota", "Camry", num_doors=4)
    car2 = Car("Honda", "Accord", num_doors=4)
    truck1 = Truck("Ford", "F-150", cargo_capacity="Short bed")
    motorcycle1 = Motorcycle("Harley-Davidson", "Sportster", has_sidecar=False)
    motorcycle2 = Motorcycle("BMW", "R 1250 GS", has_sidecar=True)

    garage = Garage([car1, car2, truck1, motorcycle1])

    garage.add_vehicle(motorcycle2)

    print(garage)
```
